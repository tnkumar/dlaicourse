# The Hello World of Deep Learning with Neural Networks

## Imports

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Define and Compile the Neural Network

Create a neural network with 1 layer, that layer with 1 neuron, and the input shape to it is just 1 value.

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Specify a loss function 'mean_squared_error' and optimizer 'sgd' - stochastic gradient descent before compiling the neural network. The learning will typically be done for a number of epochs.


```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Providing the Data

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

# Training the Neural Network

```
model.fit(xs, ys, epochs=500)
```

```
print(model.predict([10.0]))
```
